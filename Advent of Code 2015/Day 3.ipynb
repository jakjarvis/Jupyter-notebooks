```python
# Use the same input of lists - 'Dir'

# Use a counter to switch appends between two 'Coord' variables. Write them to the same 'Visits' list

# Somehow search the list for number of unique entries as before
```

```python
Coord1 = [0,0]
Coord2 = [0,0]
Counter = 1
Stops = [[0,0]]
```

```python
for char in Dir:
    x = char
    if x == "^":
        if Counter == 1:
            Coord1[0] = Coord1[0]+1
            Stops.append(Coord1.copy())
            Counter = 2
        else:
            Coord2[0] = Coord2[0]+1
            Stops.append(Coord2.copy())
            Counter = 1
    elif x == "v":
        if Counter == 1:
            Coord1[0] = Coord1[0]-1
            Stops.append(Coord1.copy())
            Counter = 2
        else:
            Coord2[0] = Coord2[0]-1
            Stops.append(Coord2.copy())
            Counter = 1
    elif x == "<":
        if Counter == 1:
            Coord1[1] = Coord1[1]-1
            Stops.append(Coord1.copy())
            Counter = 2
        else:
            Coord2[1] = Coord2[1]-1
            Stops.append(Coord2.copy())
            Counter = 1
    elif x == ">":
        if Counter == 1:
            Coord1[1] = Coord1[1]+1
            Stops.append(Coord1.copy())
            Counter = 2
        else:
            Coord2[1] = Coord2[1]+1
            Stops.append(Coord2.copy())
            Counter = 1
```
