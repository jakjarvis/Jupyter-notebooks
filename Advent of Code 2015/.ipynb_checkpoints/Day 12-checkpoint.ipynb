Santa's Accounting-Elves need help balancing the books after a recent order. Unfortunately, their accounting software uses a peculiar storage format. That's where you come in.

They have a JSON document which contains a variety of things: arrays ([1,2,3]), objects ({"a":1, "b":2}), numbers, and strings. Your first job is to simply find all of the numbers throughout the document and add them together.

For example:

- [1,2,3] and {"a":2,"b":4} both have a sum of 6.
- [[[3]]] and {"a":{"b":4},"c":-1} both have a sum of 3.
- {"a":[-1,1]} and [-1,{"a":1}] both have a sum of 0.
- [] and {} both have a sum of 0.

You will not encounter any strings containing numbers.

What is the sum of all numbers in the document?

```python
inpt = open('C:\\Users\\Jono\\Documents\\Python Scripts\\Day 12 Input.txt')

txt = inpt.read()

inpt.close()

print(txt)
```

```python
lst = list(txt)

# lst = ['[', '[', '"', 'g', 'r', 'e', 'e', '-', '8', ',', '[', '{', '"', 'e', '"', ':', '"', 'g', 'r', 'e', 'e', 'n', '"', ',', '"', 'a', '"', ':', '7', '7', ',', '"', 'd', '"', ':', '{', '"', 'c', '"', ':', '"', 'v', '9', 'o', 'l', 'e', 't', '"', ',', '"', 'a', '"', ':', '"', 'y', 'e', 'l', 'l', 'o']

print(lst)
```

```python
counter = 0

total = 0

string = ''

for x in lst:
    
    # print(x)
    
    if ord(x) >= 48 and ord(x) <=57:
        
        string = string + x
        
        if ord(lst[counter + 1]) >= 48 and ord(lst[counter + 1]) <=57:
            pass
        
        else:              
            num = int(string)
            total = total + num
            string = ''
        
        print('Total is ' + str(total))
    
    elif x == '-' and ord(lst[counter + 1]) >= 48 and ord(lst[counter + 1]) <=57:
        string = string + x
        print('Negative')
        
    counter += 1
                
print(total)  
```

```python
def loop(x, count, typ):
    
    valueX = 0
    level = count
    
    for a in x:
          
        if type(a) is dict:
            # print('Level ' + str(level) + ' element ' + str(a) + ' is a dict.\nIterating further.\n')
            output = loop(a.values(), level + 1, 'd')
            # print('Level ' + str(level) + ' element ' + str(a) + ' has value ' + str(output) + '.\nAdding ' + str(output) + ' to level ' + str(level) + ' element value ' + str(valueX) + '.')
            valueX += output
            # print('Level ' + str(level) + ' element value is now ' + str(valueX) + '.\n')

            
        elif type(a) is list or type(a) is tuple: 
            # print('Level ' + str(level) + ' element ' + str(a) + ' is a tuple or list.\nIterating further.\n')
            output = loop(a, level + 1, 'l')
            # print('Level ' + str(level) + ' element ' + str(a) + ' has value ' + str(output) + '.\nAdding ' + str(output) + ' to level ' + str(level) + ' element value ' + str(valueX) + '.')
            valueX += output
            # print('Level ' + str(level) + ' element value is now ' + str(valueX) + '.\n')
            
        else:
            if a =='red' and typ == 'd':
                valueX = 0
                # print('Level ' + str(level) + ' element ' + str(a) + ' is RED. Element ' + str(x) + ' is set to 0.\n')
                break
            elif a =='red' and typ == 'l':
                pass
            elif type(a) is int:
                # print('Level ' + str(level) + ' element ' + str(a) + ' is an integer.\nElement value ' + str(valueX) + ' is increased by ' + str(a) + '.')
                valueX += a
                # print('Level ' + str(level) + ' element value is now ' + str(valueX) + '.\n')
        
        if level == 0:
            print('\n\nTotal value is now ' + str(valueX) + '.\n\n')
    
    return valueX
```
