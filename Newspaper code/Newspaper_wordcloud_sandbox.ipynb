# Find and parse comparable news articles

Using the newspaper module as per http://theautomatic.net/2020/08/05/how-to-scrape-news-articles-with-python/

This has been performed on 3 news articles from different sources (BBC, Al Jazeera, Reuters) on a similar subject (Western countries imposing sanctions on China over the treatment of Uighurs)

```python
import newspaper as nws
```

```python
urlBBC = "https://www.bbc.com/news/world-europe-56487162"
```

```python
bbcArticle = nws.Article(urlBBC)
bbcArticle.download()
bbcArticle.parse()
# print(bbcArticle.text)
```

```python
urlAlJazeera = "https://www.aljazeera.com/news/2021/3/22/eu-rolls-out-sanctions-on-chinese-officials-over-uighur-abuses"
```

```python
alJazeeraArticle = nws.Article(urlAlJazeera)
alJazeeraArticle.download()
alJazeeraArticle.parse()
print(alJazeeraArticle.text)
```

```python
urlReuters = "https://www.reuters.com/article/us-usa-china-eu-sanctions/west-sanctions-china-over-xinjiang-abuses-beijing-hits-back-at-eu-idUSKBN2BE2MC"
```

```python
reutersArticle = nws.Article(urlReuters)
reutersArticle.download()
reutersArticle.parse()
# print(reutersArticle.text)
```

# Test the word cloud module

Using the wordcloud module and the dummy data as per https://www.datacamp.com/community/tutorials/wordcloud-python

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("archive/winemag-data-130k-v2.csv", index_col=0)
```

```python
df.head()
```

```python
print("There are {} observations and {} features in this dataset. \n".format(df.shape[0], df.shape[1]))
print("There are {} types of wine in this dataset such as {}... \n".format(len(df.variety.unique()), ", ".join(df.variety.unique()[0:5])))
print("There are {} countries producing wine in this dataset such as {}... \n".format(len(df.country.unique()),", ".join(df.country.unique()[0:5])))
```

```python
df[["country", "description", "points"]].head()
```

```python
# To make comparisons between groups of a feature, you can use groupby() and compute summary statistics.

country = df.groupby("country")
country.describe().head()
```

```python
country.mean().sort_values(by="points", ascending=False).head()
```

```python
# You can plot the number of wines by country using the plot method of Pandas DataFrame and Matplotlib.

plt.figure(figsize=(15,10))
country.size().sort_values(ascending=False).plot.bar()
plt.xticks(rotation=50)
plt.xlabel("Country of Origin")
plt.ylabel("Number of Wines")
plt.show()
```

```python
# You can examine the function with ?
?WordCloud
```

```python
# So let's start with a simple example: using the first observation description as the input for the wordcloud

text = df.description[0]

wordcloud = WordCloud().generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
