# Initial Analysis

Using data previously created in "Data Compilation", file printed to csv according to last date stamp

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("30-04-2021.csv", index_col = ["date", "yahoo_index"])
data.head()
```

```python
data.info()
```

```python
data_subset1 = data.iloc[:, [1, 3, 4, 6, 10, 16]]
```

```python
data_subset1.loc[:, "Close"].argmax()
```

```python
data_subset1.iloc[data_subset1.loc[:, "Close"].argmax()]
```

```python
data_subset1.corr()
```

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn")
data_subset1.loc[:, "Close"].unstack().plot(figsize = (15, 8 ), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```

```python
data_subset1 = data_subset1.reset_index()[data_subset1.reset_index().yahoo_index != "^VIX"]
data_subset1.set_index(["date", "yahoo_index"], inplace = True)
```
