```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
order_book = pd.DataFrame(data = {"name":["Rolls-Royce", "Rolls-Royce", "Shionogi", "Euro Stoxx"], "yahoo-code":["RR.L", "RR.L", "4507.T", "EXW1.DE"], "volume":[2000, 1500, 50, 90], "purchase-date":["2020-11-24", "25-03-2021", "01-03-2021", "05-05-2021"], "purchase-price":[132.9, 120.6, 42.74, 40.56], "currency":["GBP", "GBP", "JPY", "EUR"]})
order_book["total-purchase"] = order_book.loc[:, ["volume", "purchase-price"]].product(axis=1)
order_book
```

```python
portfolio = pd.DataFrame(columns = ["name", "yahoo-code", "volume", "average-price", "currency", "total-purchase"])
for stock_name in order_book["name"].unique().tolist():
    stock_info = order_book[order_book["name"] == stock_name].iloc[0]
    stock_sum = order_book[order_book["name"] == stock_name].sum(axis = 0)
    stock_mean = order_book[order_book["name"] == stock_name].mean(axis = 0)
    portfolio = portfolio.append({"name":stock_name, "yahoo-code":stock_info["yahoo-code"], "volume":stock_sum["volume"], "average-price":stock_mean["purchase-price"], "currency":stock_info["currency"], "total-purchase":stock_sum["total-purchase"]}, ignore_index=True)
portfolio
```

```python
ticker = order_book["yahoo-code"].tolist()
stocks = yf.download(ticker, start = "2020-01-01", end = "2021-04-30")
```

```python
exchange = ["EURGBP=X", "EURJPY=X"]
exchange_rates = yf.download(exchange, start = "2020-01-01", end = "2021-04-30")
```

```python
close = stocks["Close"].copy()
rates = exchange_rates["Close"]
close["RR.L"] = close["RR.L"] / rates["EURGBP=X"]
close["4507.T"] = close["4507.T"] / rates["EURJPY=X"]
close
```

```python
order_book.set_index("yahoo-code", inplace = True)
portfolio.set_index("yahoo-code", inplace = True)
```

```python
stock = "4507.T"
portfolio.loc[stock, "average-price"]
```

```python
for stock in close.columns.tolist():
    close.loc[:, stock] = close.loc[:, stock] / portfolio.loc[stock, "average-price"]
```

```python
print(close[close.loc[:, 'RR.L'] < 0.1])
close.loc["2021-04-13", "RR.L"] = np.nan
```

```python
plt.style.use("seaborn")
close.plot(figsize = (15, 8 ), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```
