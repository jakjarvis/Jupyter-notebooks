#### According to Quora
"Equestrian and Sailing events are typically mixed gender"<br>

Most of the more recent records fall into these catagories. Some older ones are probably due to data integrity - we can revisit this later if needs be.

```python
summer.Gender[summer.Sport == "Equestrian"] = "Mixed"
summer.Gender[summer.Sport == "Sailing"] = "Mixed"
summer[summer.Sport == "Sailing"]
```

```python
summer_count = summer.loc[:, ["Year", "Discipline", "Event", "Gender", "Medal"]].groupby(by = ["Year", "Discipline", "Event", "Gender"]).count().reset_index()
summer_count[summer_count.Discipline == 'Dressage']
```

```python
# Check if any events had different medals counts in different years
print(summer_count.loc[:, ["Discipline", "Event", "Gender", "Medal"]].duplicated(keep = False).value_counts())
# Not all are unique
```

```python
singles = summer_count[summer_count.Medal < 6].loc[:, ["Discipline", "Event", "Medal"]]
unique_singles = singles[singles.duplicated(keep = "first")]
unique_singles
```

```python
multiples = summer_count[summer_count.Medal > 5].loc[:, ["Discipline", "Event", "Medal"]]
unique_multiples = multiples[multiples.duplicated(keep = "first")]
unique_multiples
```

```python
unique_singles.merge(unique_multiples, how = "inner", on = ["Discipline", "Event"])
```

Thus all sports are unique to only one list.

```python
unique_singles.drop(columns = "Medal", inplace = True)
unique_singles = unique_singles[unique_singles.duplicated(keep = "first")]

unique_multiples.drop(columns = "Medal", inplace = True)
unique_multiples = unique_multiples[unique_multiples.duplicated(keep = "first")]
```

```python
unique_singles
```

```python
unique_multiples
```

```python
summer_edited = pd.read_csv("summer_clean.csv")
summer_edited.Gender[summer_edited.Sport == "Equestrian"] = "Mixed"
summer_edited.Gender[summer_edited.Sport == "Sailing"] = "Mixed"
```

```python
summer_edited
```

```python
summer_edited = summer_edited.merge(unique_multiples, how = "outer", on = ["Discipline", "Event"], indicator = True).drop_duplicates()
```

```python
summer_edited = summer_edited.drop(columns = ["Athlete", "Unnamed: 0"]).drop_duplicates().reset_index().drop(columns = "index")
```

```python
summer_edited
```

```python
summer_1976 = summer_edited[summer_edited.Year == 1976].loc[:, ["Country", "Medal"]]
summer_1976["Count"] = 1
```

```python
summer_1976 = summer_1976.groupby(by = ["Country", "Medal"]).count().unstack().fillna(0)
```

```python
summer_1976.columns = summer_1976.columns.droplevel()
```

```python
summer_1976 = summer_1976[["Gold", "Silver", "Bronze"]].sort_values(by = ["Gold", "Silver", "Bronze"], ascending = False)
summer_1976.sum()
```

```python
summer_1996 = summer_edited[summer_edited.Year == 1996].loc[:, ["Country", "Medal"]]
summer_1996["Count"] = 1
summer_1996 = summer_1996.groupby(by = ["Country", "Medal"]).count().unstack().fillna(0)
summer_1996.columns = summer_1996.columns.droplevel()
summer_1996 = summer_1996[["Gold", "Silver", "Bronze"]].sort_values(by = ["Gold", "Silver", "Bronze"], ascending = False)
pd.set_option('display.max_rows', 500)
summer_1996
```
