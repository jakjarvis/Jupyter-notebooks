### Ranking DataFrames with rank()

```python
import pandas as pd
```

```python
sales = pd.Series([15, 32, 45, 21, 55, 15, 0],  index = ["Mo", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
```

```python
sales
```

```python
sales.sort_values(ascending = False)
```

```python
sales.rank(ascending=False, method = "min").sort_values(ascending = True) # method says how you assign to shared values
```

```python
sales.rank(ascending=False, method = "min", pct=True).sort_values()
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.fare.rank(ascending = False)
```

```python
titanic["fare_rank"] = titanic.fare.rank(ascending = False, method="min")
```

```python
titanic.head()
```

```python
titanic.sort_values("fare", ascending= False)
```

```python
titanic.drop(columns = "fare_rank", inplace= True)
```

### apply(), map() and applymap()

```python
import pandas as pd
```

```python
sales = pd.read_csv("sales.csv", index_col = 0)
```

```python
sales
```

```python
sales.info()
```

```python
sales.min(axis = 0)
```

```python
sales.min(axis = 1)
```

```python
def range(series):
    return series.max() - series.min()
```

```python
sales.apply(range, axis = 1)
```

```python
sales.apply(lambda x: x.max() - x.min(), axis = 0) # lamda means we will define the series next
```

```python
summer = pd.read_csv("summer.csv") 
```

```python
summer.head()
```

```python
summer.Athlete.apply(lambda x: x[0]) # .apply applies the function to the specified dataframe
```

```python
summer.Athlete.map(lambda x: x[0]) # map works on every element of a series (the same in this case as apply)
```

```python
summer.iloc[:,1:3].applymap(lambda x: x[0]) # applymap applies to all elements of a dataframe
```

```python
sales.applymap(lambda x: 0.4*x-5)
```

```python
sales*0.4-5
```
