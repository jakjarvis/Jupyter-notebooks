# Indexing and Slicing with reindex()

```python
summer = pd.read_csv("summer.csv") # removing the athlete index we previously added, just using a range index
```

```python
summer.loc[[0, 5, 30000], ["Athlete", "Medal"]]
```

```python
summer.loc[[0, 5, 30000, 40000], ["Athlete", "Medal"]] # 40000 is too high, not in list
```

```python
summer.reindex(index = [0, 5, 30000, 40000], columns = ["Athlete", "Medal", "Age"]) # prevents the error if data not there
```

```python
summer.reindex(index = [0, 5, 30000, 40000], columns = ["Athlete", "Medal", "Age"])
```

```python
summer = pd.read_csv("summer.csv", index_col = "Athlete") # making the index the athlete again
```

```python
summer.reindex(columns = ["Medal", "Age"])
```

```python
summer.reindex(index = ["PHELPS, Michael"], columns = ["Medal", "Age"]) # doesn't work with non unique index
```
